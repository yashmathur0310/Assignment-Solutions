```python
df=df[['source_name', 'destination_name','firm','speed actual','cc1 lcu amount','cc1 lcu fx rate']]
df_india=df[(df['destination_name'] == 'India')]
df_india.drop(['destination_name'],axis=1,inplace=True)
df_india=df_india[df_india['firm'].isin(df_india['firm'].value_counts()[df_india['firm'].value_counts() > 10].index)]
```

```python
df_dummies=pd.get_dummies(df_india, columns=['source_name', 'speed actual'])
df_dummies
```

```python
x=df_dummies.drop(['firm'],axis=1).values
y=df_dummies['firm'].values
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42,stratify=y)
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
from sklearn.metrics import *
```

```python
print("Accuracy on the training dataset:",model.score(x_train,y_train))
print("Accuracy on the testing dataset:",model.score(x_test,y_test))
```

```python
print(classification_report(y_test,model.predict(x_test)))
```

```python
recall_score(y_test,model.predict(x_test),average='weighted')
```
